```python
import zipfile
from docx import Document

def read_file(file_path):
    """
    Extracts text from a file based on its format (PDF or DOCX).

    Args:
        file_path (str): Path to the file.

    Returns:
        str: Extracted text from the file.
    """
    if file_path.lower().endswith('.pdf'):
        # If PDF, extract text
        with open(file_path, 'rb') as file:
            reader = PyPDF2.PdfReader(file)
            text = "".join(page.extract_text() for page in reader.pages if page.extract_text())
        return text
    elif file_path.lower().endswith('.docx'):
        try:
            # Try to open the DOCX file
            doc = Document(file_path)
            text = "\n".join(paragraph.text for paragraph in doc.paragraphs)
            return text
        except zipfile.BadZipFile:
            # Handle invalid or corrupted DOCX file
            raise ValueError("The provided .docx file seems to be corrupted or invalid.")
    else:
        raise ValueError("Unsupported file format. Only 'pdf' and 'docx' are supported.")
```

```python
def calculate_similarity(jd_embedding, resume_embedding):
    """
    Calculate cosine similarity between job description and resume text embeddings.
    """
    similarity_score = cosine_similarity(jd_embedding, resume_embedding)[0][0] * 100  # Get percentage score

    # Print similarity reasoning
    print("\nSimilarity Score:", f"{similarity_score:.2f}%")
    print("\nReasoning:")

    if similarity_score > 60:
        print("The high similarity indicates a strong alignment between the job description and the resume.")
    elif similarity_score > 30:
        print("The moderate similarity suggests some overlap between the resume and the job description.")
    else:
        print("The low similarity indicates a weak alignment between the resume and the job description.")

    return similarity_score
```

```python
def main(jd_file_path, resume_file_path):
    try:
        # Step 1: Extract text from the files (resume and job description)
        jd_text = read_file("/content/JOB DESCRIPTION.docx")  # Path to job description file
        resume_text = read_file("/content/PavanK_Resume.docx")  # Path to resume file
        # Step 2: Get embeddings for JD and resume using pre-trained model
        jd_embedding = model.encode([jd_text])  # Get embedding for job description
        resume_embedding = model.encode([resume_text])  # Get embedding for resume

        # Step 3: Calculate the similarity score and provide reasoning
        similarity_score = calculate_similarity(jd_embedding, resume_embedding)

        return similarity_score  # Return the similarity score

    except Exception as e:
        print(f"Error: {e}")
        return None

```
